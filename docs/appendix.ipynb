## Gaußsche Integral

````{prf:theorem} Gaußsche Integral
```{math}
:label: gaussintegraleq
\int_{-\infty}^{\infty}{e^{-a x^2}} dx = \sqrt{\frac{\pi}{a}}
```
````

````{prf:proof}
:label: gaussintegralproof

Zu lösen ist das Integral der Gaußschen Verteilungsfunktion.

```{math}
\int_{-\infty}^{\infty}{e^{-a x^2}} dx
```

Erst quadrieren wir die Gleichung

```{math}
\Bigl(\int_{-\infty}^{\infty}{e^{-a x^2}} dx\Bigr)^2 = \int_{-\infty}^{\infty}{e^{-a x^2}} dx \int_{-\infty}^{\infty}{e^{-a y^2}} dy
```

Das Integral können wir zusammenfassen und wir erhalten
```{math}
\int_{-\infty}^{\infty} \int_{-\infty}^{\infty}{e^{-a (x^2 + y^2)}} dx dy
```

Dann können wir in Zylinderkoordinaten übergehen mit $r = \sqrt{x^2 + y^2}$ und $dx dy = r dr d\theta$.
```{math}
\int_{0}^{2 \pi}{ d\phi}\int_{0}^{\infty}{e^{-a r^2}} r dr
```

Nach Integration erhält man
```{math}
\int_{0}^{2 \pi}{ d\theta}\int_{-\infty}^{\infty}{e^{-a r^2}} r dr = 2 \pi \biggl[-\frac{1}{2 a}e^{-a r^2}\biggr]_{0}^{\infty} = \frac{\pi}{a}
```

Das Gauß-Integral kann man folglich exakt auswerten als

```{math}
:label: gaussintegraleq_final
\int_{-\infty}^{\infty}{e^{-a x^2}} dx = \sqrt{\frac{\pi}{a}}
```
````

## Binominal Distribution

Let's say we roll a dice $n$ times. Now we want to find the probability of 
getting the six face upwards $k$ times. We assume an unrigged dice so the 
probability of getting a six is $p_{six} = \frac{1}{6}$ and the probability 
of not getting a six is $\overline{p_{six}} = 1 - p_{six} = \frac{5}{6}$. 
One first guess on the probability to get $k$ sixes out of $n$ throws is 
$p_{six}^k (1 - p_{six})^{(n-k)}$. However there we disregard that are multiple 
possibilities for getting $k$ sixes. We could either get the $k$ sixes 
immediately at our first throws or at the very end or at any other random 
distribution. There the binomial coefficient stems from. The binomial 
coefficient describes how many possibilities there are to arrange 
the $k$ sixes in the total amount of throws $n$. For the first six we have 
$n$ possibilities, for the second one $n-1$ and so on until the last six 
$n-k$. In total $n(n-1)(n-2) \dots (n-k) = \frac{n!}{(n-k)!}$ possibilities. However as the sixes are indistinguishable we have to divide this by $k!$ to 
get the binomial coefficient $\frac{n!}{k!(n-k)!}$. So 
the probability of getting $k$ sixes with $n$ dice throws is 

```{math}
P(k, n) = \frac{n!}{k!(n-k)!} p_{six}^k (1 - p_{six})^{(n-k)}
```

For distribution looks as follows for a total number of $n=40$ throws and 
probabilities of $p=1/6$ and $p=3/6$.

`````python

import matplotlib.pyplot as plt
import numpy as np
import scipy

fig, axes = plt.subplots()

def probability(k, n, p):
    return scipy.special.binom(n, k)*(p**k)*((1-p)**(n-k))


axes.bar(np.arange(40), probability(k=np.arange(40), n=40, p=1/6), label=r'$N=40, p=1/6$')
axes.bar(np.arange(40), probability(k=np.arange(40), n=40, p=3/6), label=r'$N=40, p=3/6$')
axes.legend()
plt.xlabel('Number of occurences')
plt.ylabel('Probability')
plt.show()
`````
